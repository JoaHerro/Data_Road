# How weel do jobs and skills pay for Data Analysts?

## Methodology

 1. Evaluate median salary for top 6 data jobs
 2. Find median salary per skill for Data Analysts
 3. Visualize for highest paying skills and most demanded skills

```python
# Importing Libraries
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt
import seaborn as sns

#! pip install adjustText
#from adjustText import adjust_text


# loading data
dataset = load_dataset("lukebarousse/data_jobs")
df = dataset['train'].to_pandas()

# Data cleaning
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df_US = df[(df['job_country'] == 'United States')].dropna(subset=['salary_year_avg'])

df_US
```

```python
job_titles = df_US['job_title_short'].value_counts().index[:6].to_list()
```

```python
job_titles
```

```python
df_US_top6 = df_US[df_US['job_title_short'].isin(job_titles)]

df_US_top6
```

```python
job_ordered = df_US_top6.groupby('job_title_short')['salary_year_avg'].median().sort_values(ascending=False).index
```

```python
sns.boxplot(data= df_US_top6, x='salary_year_avg', y='job_title_short', order=job_ordered)
sns.set_theme(style='ticks')

plt.title('Salary Distribution in the United States')
plt.xlabel('Yearly Salary in USD')
plt.ylabel('')
ax = plt.gca()
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, pos: f'${int (x/1000)}K'))
plt.xlim(0,600_000)

plt.show()
```
