```python
import pandas as pd
from datasets import load_dataset

#Loading Data
dataset = load_dataset("lukebarousse/data_jobs")
df = dataset['train'].to_pandas()

#Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df.describe()
```

```python
df.info()
```

```python
df.count()
```

```python
df.describe()
```

```python
# Min value from a column with integers   df.min()

df['salary_year_avg'].min()
```

```python
# Index in the Data Frame (df) of mininum value

min_salary = df['salary_year_avg'].idxmin()
max_salary = df['salary_year_avg'].idxmax()
```

```python
# Access to specific index from a Data Frame

df.iloc[min_salary]
```

```python
df.iloc[max_salary]
```

```python
# df.unique() use to get unique values from a Series (single columns or Data frames) similar to Sets

df['job_title_short'].unique()

```

```python
# df.value_counts() returns the counts of unique values in a Pandas Series

df['job_title_short'].value_counts()
```

```python
df.groupby('job_title_short')['salary_year_avg'].min()
```

```python
#To gather them with multiple columns use ["colum", "colum"] brakets separeted by commas

df.groupby('job_title_short')[['salary_year_avg', 'salary_hour_avg']].median()
```

```python
# For multiple methods use df,agg(["min", "max", "mean"])

df.groupby(['job_title_short', 'job_country'])[['salary_year_avg', 'salary_hour_avg']].agg(["min", "max", "mean"])
```

# Matplotlib: Labeling

plt.title('Wazxup, Data Nerd!')

```python
job_title_short = df.job_title_short.value_counts(ascending=False)
plt.bar(job_title_short.index, job_title_short)
plt.title('Most common job titles')
plt.ylabel('Job Title')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
job_title_short.plot(kind='bar')
plt.title('Most common job titles')
plt.ylabel('Job Title')
plt.xlabel('')
plt.xticks(rotation=45, ha='right')
plt.show()
```
