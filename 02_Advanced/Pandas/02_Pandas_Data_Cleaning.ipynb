```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# loading data
dataset = load_dataset("lukebarousse/data_jobs")
df = dataset['train'].to_pandas()

# Data cleaning
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df.info()
```

```python
df.describe()
```

```python
df.loc[:10,'salary_year_avg':'salary_hour_avg']
```

```python
salary_year_avg = df['salary_year_avg'].median()
```

```python
salary_hour_avg = df['salary_hour_avg'].median()
```

```python
df_filled = df
```

```python
df_filled['salary_year_avg'] = df_filled['salary_year_avg'].fillna(salary_year_avg)
df_filled['salary_hour_avg'] = df_filled['salary_hour_avg'].fillna(salary_hour_avg)
```

```python
df_filled['salary_year_avg']
```

```python
df_filled.loc[:10,'salary_year_avg': 'salary_hour_avg']
```

```python
df_unique = df_filled

df_unique = df_unique.drop_duplicates()
```

```python
df_unique.describe()
```

```python
df_unique.info()
```

```python
df_unique = df_unique.drop_duplicates(subset=['job_title', 'company_name'])
```
